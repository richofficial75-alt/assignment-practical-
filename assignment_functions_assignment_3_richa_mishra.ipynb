4. What is the purpose of the `return` statement in a Python function?

The return statement in a Python function serves two main purposes:

Exiting the Function: It immediately terminates the execution of the function.

Sending a Value Back: It specifies the value (or multiple values) that the function should output back to the caller. If a function reaches the end without a return statement, it implicitly returns the special value None.

Example:



def calculate_area(radius):

    if radius < 0:
       
        return None # Exits and returns None for invalid input

    return area     # Returns the calculated area

area2 = calculate_area(-5)

print(f"Area 1: {area1}")

print(f"Area 2: {area2}")

# Area 1: 314.0

5.What are iterators in Python and how do they differ from iterables?

An iterable is any object in Python that can be "iterated over," meaning it can return its members one at a time. Examples include lists, tuples, strings, and dictionaries. An iterable has the __iter__ method.

An iterator is an object that represents a stream of data. It is the object that actually does the iteration. It must implement the __iter__ method (which returns itself) and the __next__ method, which returns the next item from the container. When there are no more items, it raises a StopIteration exception.

Example:

# Use the iterator to get the next element

print(next(my_iterator))
# Output:
# 10
# 20
# If we call next() again after 30, it raises StopIteration.

6.Explain the concept of generators in Python and how they are defined.

A generator is a simple and fast way to create iterators. Unlike regular functions, which compute a result and return it all at once, a generator function yields a sequence of results one at a time, pausing its execution state between calls. This means it doesn't store the entire result sequence in memory.

Generators are defined just like regular functions, but instead of the return statement, they use the yield statement.

Example (Generator Function):



def count_up_to(max_val):
    n = 1
    while n <= max_val:
        yield n # Pauses and returns 'n', saves state for next call
        n += 1

# Calling the function returns a generator object
counter = count_up_to(3)

print(next(counter))

print(next(counter))
# Output:
# 1
# 2

 7.What are the advantages of using generators over regular functions?

 The main advantages of using generators over regular functions that return a list are:

Memory Efficiency: They are lazy and compute values only when requested, yielding one item at a time. This is crucial for working with very large or infinite sequences, as the entire sequence isn't stored in memory.

Performance: They start producing results immediately and don't wait for the entire sequence to be generated.

Simplicity: The code to write a generator is often more concise than writing a custom iterator class.

Example (Memory Efficiency):
A function generating a list of a billion numbers would consume huge amounts of memory, whereas a generator only holds the current state and the next number to be generated.



# Regular function (creates a list of 1000000 numbers in memory)
def create_big_list():

    return list(range(1000000))

# Generator function (creates numbers one-by-one, low memory usage)
def create_big_generator():
    for i in range(1000000):
        yield i

# Iterating over the generator is memory efficient
for num in create_big_generator():

    if num == 10:
        break
    # Only the current state is stored

8.What is a lambda function in Python and when is it typically used?

A lambda function (or anonymous function) is a small, restricted function defined in a single line using the lambda keyword. It can take any number of arguments, but it can only have one expression, and the result of this expression is implicitly returned. They are restricted because they cannot contain statements like if, for, or explicit return.

They are typically used when a small, throw-away function is required for a short period, often as an argument to higher-order functions like map(), filter(), or sorted().

Example:



# A lambda function to calculate the square of a number
square = lambda x: x * x
print(square(9))
# Output: 81

# Usage with sorted() to sort a list of tuples by the second element
data = [('apple', 3), ('banana', 1), ('cherry', 2)]
sorted_data = sorted(data, key=lambda item: item[1])
print(sorted_data)
# Output: [('banana', 1), ('cherry', 2), ('apple', 3)]

10.What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

These are built-in higher-order functions that process iterables but have distinct purposes:

map()

Purpose: Transformation (applies a function to every item).

Syntax: map(function, iterable)

Output: A map object (an iterator) containing the results of the transformation for each item.

Example:



numbers = [1, 2, 3]
# Squares every number: 1*1, 2*2, 3*3
mapped = list(map(lambda x: x * x, numbers))
# Output: [1, 4, 9]
filter()

Purpose: Selection (constructs an iterator from elements for which a function returns True).

Syntax: filter(function, iterable)

Output: A filter object (an iterator) containing only the selected elements.

Example:



numbers = [1, 2, 3, 4]
# Keeps only even numbers
filtered = list(filter(lambda x: x % 2 == 0, numbers))

Output: [2, 4]

reduce()

Purpose: Aggregation (applies a function cumulatively to the items to reduce the iterable to a single cumulative value).

Syntax: reduce(function, iterable[, initial])

Output: The single calculated value.

Note: It must be imported from the functools module (from functools import reduce).

Example:

from functools import reduce
numbers = [1, 2, 3, 4]
# Calculates the sum: (((1+2)+3)+4)
reduced = reduce(lambda acc, x: acc + x, numbers)
# Output: 10

```python
#1.Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.
def sum_even_numbers(numbers):
    """
    Calculates the sum of all even numbers in a list.
    """
    total_sum = 0
    for number in numbers:
        if number % 2 == 0:
            total_sum += number
    return total_sum

# Example
my_list = [10, 5, 2, 8, 3, 1]
result = sum_even_numbers(my_list)
print(f"Sum of even numbers in {my_list}: {result}")
```

```python
#4. Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(number):
    """
    Checks if a given number (assumed to be between 1 and 200) is prime.
    """
    if number <= 1:
        return False
    if number <= 3:
        return True
    if number % 2 == 0 or number % 3 == 0:
        return False

    # Check for factors up to the square root of the number
    i = 5
    while i * i <= number:
        if number % i == 0 or number % (i + 2) == 0:
            return False
        i += 6
    return True

# Example
num1 = 17
num2 = 15
print(f"Is {num1} prime? {is_prime(num1)}")
print(f"Is {num2} prime? {is_prime(num2)}")
```

```python
#5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
class FibonacciIterator:
    """
    An iterator that generates the Fibonacci sequence up to a specified number of terms.
    """
    def __init__(self, terms):
        self.terms = terms
        self.count = 0
        self.a = 0
        self.b = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.terms:
            raise StopIteration

        if self.count == 0:
            self.count += 1
            return 0 # First term

        if self.count == 1:
            self.count += 1
            return 1 # Second term

        # For subsequent terms
        result = self.a + self.b
        self.a = self.b
        self.b = result
        self.count += 1
        return result

# Example
fib_seq = FibonacciIterator(7)
print("Fibonacci Sequence (7 terms):", end=" ")
for num in fib_seq:
    print(num, end=" ")
```

```python
#6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def powers_of_two(max_exponent):
    """
    A generator function that yields the powers of 2 up to a given exponent.
    """
    for exp in range(max_exponent + 1):
        yield 2 ** exp

# Example
pow_gen = powers_of_two(5)
print("Powers of 2 up to exponent 5:", end=" ")
print(list(pow_gen))
```

```python
#7. Implement a generator function that reads a file line by line and yields each line as a string.
import os

def file_line_generator(filepath):
    """
    A generator function that reads a file line by line and yields each line.
    """
    try:
        with open(filepath, 'r') as file:
            for line in file:
                yield line.rstrip('\n') # Yield the line, removing the newline character
    except FileNotFoundError:
        print(f"Error: The file '{filepath}' was not found.")

# Example Setup: Create a temporary file
test_file = "temp_lines.txt"
with open(test_file, 'w') as f:
    f.write("First line.\n")
    f.write("Second line.\n")
    f.write("Third line.")

# Example Usage
print("Reading file contents line by line:")
line_gen = file_line_generator(test_file)
for line in line_gen:
    print(f"-> {line}")
# Clean up the file
os.remove(test_file)
```

```python
#9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
def celsius_to_fahrenheit(celsius):
    """Converts a temperature from Celsius to Fahrenheit."""
    return (celsius * 9/5) + 32

celsius_temps = [0, 25, 100, -10]

# Use map() to apply the conversion function to every element
fahrenheit_map = map(celsius_to_fahrenheit, celsius_temps)

# Convert the map object to a list
fahrenheit_temps = list(fahrenheit_map)

print(f"Celsius Temperatures: {celsius_temps}")
print(f"Fahrenheit Temperatures: {fahrenheit_temps}")
```
